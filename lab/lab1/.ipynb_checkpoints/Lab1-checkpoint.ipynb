Import libraries and open image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

Image_path = "cat.png"
Image = cv2.imread(Image_path,0)
Size =Image.shape
Size
```

Initialise max/min gray value pixels numpy array and 0, 255 array that smilier to image size

```python
a = 0
b = 255
c = np.min(Image)
d = np.max(Image)
```

Contrast Stretching

```python
O= np.trunc((Image - c)*((b-a)/(d-c))+a)
cv2.imwrite("O.png",O)
```

Calculate intensity histogram of input image

```python
h = [np.count_nonzero(Image == i) for i in range(256)]
plt.bar(range(256),h)
plt.title("Intensity histogram for " + Image_path)
plt.xlabel("gray value")
plt.ylabel("number of pixels value")
plt.show()
```
